```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.offline import plot
import matplotlib.pyplot as plt
import datetime as dt
from pycoingecko import CoinGeckoAPI
from mplfinance.original_flavor import candlestick2_ohlc
```

```python
# target_currencies = ['usd','eur']
target_currencies = ['usd','idr','twd','eur','rub','krw','jpy','cny']
```

```python
data = ['id', 'symbol', 'name', 'current_price', 'market_cap', 'market_cap_rank', 'total_volume', 'market_cap_change_percentage_24h', 'price_change_percentage_1h_in_currency', 'price_change_percentage_24h_in_currency', 'price_change_percentage_7d_in_currency','curr']
```

```python
cg = CoinGeckoAPI()
df = pd.DataFrame(columns=data)
```

```python
for currencies in target_currencies:
  listings = cg.get_coins_markets(vs_currency=currencies,per_page = 100,price_change_percentage='1h,24h,7d')
  frame = []
  for listing in listings:
    new_line =[]
    
    for j in data[:-1]:
      new_line.append(listing[j])
    new_line.append(currencies)
    frame.append(new_line)
    
  df = df.append(pd.DataFrame(frame,columns=data))
```
